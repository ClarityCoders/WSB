```
import csv
rows = []
```

## Pick your File.

```
filename = "AMC.csv"
```

```
with open(filename, 'r') as csvfile: 
    # creating a csv reader object 
    csvreader = csv.reader(csvfile)

    for row in csvreader: 
        rows.append(row)
```

## Create a dictionary of days to count messages per day.

```
dataDict = {}

for row in rows:
    if row[2].strip() in dataDict.keys():
        dataDict[row[2].strip()] += 1
    else:
        dataDict[row[2].strip()] = 0

print(dataDict)
```

```
f = open(f'PerDay-{filename}', "w")
for i in range(1, 30):
    if str(i) in dataDict.keys():
        f.write(f"{i},{dataDict[str(i)]}\n")
    else:
        f.write(f"{i},0\n")

f.close()
```
